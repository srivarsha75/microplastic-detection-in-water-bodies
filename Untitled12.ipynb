```python
# STEP 1: Install Required Libraries
!pip install -q transformers torchvision matplotlib torch_geometric

# STEP 2: Imports
import os, glob, zipfile, random
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torch import nn
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset
import torchvision.transforms as transforms
import torchvision
from torchvision import datasets
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from transformers import ViTForImageClassification
import torch.nn.functional as F
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score


# STEP 3: Upload and Extract Dataset
from google.colab import files
uploaded = files.upload()  # Upload train.zip
with zipfile.ZipFile("train.zip", 'r') as zip_ref:
    zip_ref.extractall("train_data")

image_dir = "train_data/train"
print("Path exists:", os.path.exists(image_dir))

# STEP 4: Transforms
transform = transforms.Compose([
    transforms.Resize(32),
    transforms.CenterCrop(32),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# STEP 5: DCGAN Modules (Generator and Discriminator)
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(100, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256), nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128), nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64), nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, input):
        return self.main(input)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.main = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.BatchNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 1, 4, 1, 0), nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)

# Function to Generate and Save Images
def generate_and_save_images(generator, num_images=500, save_dir="synthetic_images"):
    class_dir = os.path.join(save_dir, "class0")
    os.makedirs(class_dir, exist_ok=True)
    generator.eval()
    z = torch.randn(num_images, 100, 1, 1, device=device)
    with torch.no_grad():
        fake_images = generator(z).detach().cpu()
    for i, img_tensor in enumerate(fake_images):
        img = transforms.ToPILImage()(img_tensor)
        img.save(os.path.join(class_dir, f"synthetic_{i}.png"))
    print(f"Saved {num_images} synthetic images to '{class_dir}'")

    # Show 8 samples
    grid = torchvision.utils.make_grid(fake_images[:8], nrow=4, normalize=True)
    plt.figure(figsize=(8, 4))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')
    plt.title("Sample Synthetic Images (DCGAN)")
    plt.show()

# Function to Visualize Progress
def visualize_progress(epoch, generator, fixed_noise, device):
    generator.eval()
    with torch.no_grad():
        fake_images = generator(fixed_noise.to(device)).detach().cpu()
    grid = torchvision.utils.make_grid(fake_images[:8], nrow=4, normalize=True)
    plt.figure(figsize=(8, 4))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')
    plt.title(f"Generated Images at Epoch {epoch}")
    plt.show()

# STEP 6: Prepare Real Dataset for GAN
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
real_dataset = datasets.ImageFolder("train_data", transform=transform)
gan_loader = DataLoader(real_dataset, batch_size=64, shuffle=True)

# STEP 7: Train DCGAN (100 epochs, with progress visualization)
generator = Generator().to(device)
discriminator = Discriminator().to(device)
criterion = nn.BCELoss()
opt_G = torch.optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
opt_D = torch.optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))

epochs = 100
fixed_noise = torch.randn(64, 100, 1, 1, device=device)

for epoch in range(epochs):
    for i, (real_images, _) in enumerate(gan_loader):
        real_images = real_images.to(device)
        b_size = real_images.size(0)
        real_labels = torch.ones(b_size, device=device)
        fake_labels = torch.zeros(b_size, device=device)

        z = torch.randn(b_size, 100, 1, 1, device=device)
        fake_images = generator(z)

        # Train Discriminator
        real_loss = criterion(discriminator(real_images).view(-1), real_labels)
        fake_loss = criterion(discriminator(fake_images.detach()).view(-1), fake_labels)
        d_loss = real_loss + fake_loss
        opt_D.zero_grad(); d_loss.backward(); opt_D.step()

        # Train Generator
        output = discriminator(fake_images).view(-1)
        g_loss = criterion(output, real_labels)
        opt_G.zero_grad(); g_loss.backward(); opt_G.step()

    print(f"Epoch {epoch+1}/{epochs}, D Loss: {d_loss.item():.4f}, G Loss: {g_loss.item():.4f}")

    if epoch % 10 == 0:
        visualize_progress(epoch, generator, fixed_noise, device)

# STEP 8: Generate Synthetic Images
generate_and_save_images(generator, num_images=500)

# STEP 9: Load Real + Synthetic Data
synthetic_dataset = datasets.ImageFolder("synthetic_images", transform=transform)
full_dataset = ConcatDataset([real_dataset, synthetic_dataset])
print("Total images (real + synthetic):", len(full_dataset))

train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_ds, test_ds = random_split(full_dataset, [train_size, test_size])
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False)

# STEP 10: Show Mixed Batch
def show_batch(data_loader):
    images, _ = next(iter(data_loader))
    grid = torchvision.utils.make_grid(images[:8], nrow=4, normalize=True)
    plt.figure(figsize=(10, 5))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')
    plt.show()

show_batch(train_loader)

# STEP 11: ViT Feature Extraction
vit_model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224-in21k')
vit_model.to(device)

vit_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

class ViTDataset(Dataset):
    def __init__(self, dataset):
        self.dataset = dataset
        self.transform = vit_transform

    def __len__(self): return len(self.dataset)

    def __getitem__(self, idx):
        img, _ = self.dataset[idx]
        img = self.transform(transforms.ToPILImage()(img))
        return img

vit_loader = DataLoader(ViTDataset(train_ds), batch_size=32, shuffle=False)

all_features = []
for images in vit_loader:
    images = images.to(device)
    with torch.no_grad():
        outputs = vit_model(images).logits
    all_features.append(outputs.cpu())
all_features = torch.cat(all_features, dim=0)

# STEP 12: KMeans + GCN
features_numpy = all_features.numpy()
kmeans = KMeans(n_clusters=4)
pseudo_labels = kmeans.fit_predict(features_numpy)
print("Silhouette Score:", silhouette_score(features_numpy, pseudo_labels))
pseudo_labels = torch.tensor(pseudo_labels, dtype=torch.long)

def build_graph_from_embeddings(embeddings, k=5):
    nbrs = NearestNeighbors(n_neighbors=k+1, metric='cosine').fit(embeddings)
    _, indices = nbrs.kneighbors(embeddings)
    edge_index = []
    for i in range(len(embeddings)):
        for j in indices[i][1:]:
            edge_index.append([i, j])
    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    return Data(x=torch.tensor(embeddings, dtype=torch.float32), edge_index=edge_index)

def create_gcn(in_channels, hidden_channels, out_channels):
    class GCN(nn.Module):
        def __init__(self):
            super().__init__()
            self.conv1 = GCNConv(in_channels, hidden_channels)
            self.conv2 = GCNConv(hidden_channels, out_channels)

        def forward(self, x, edge_index):
            x = self.conv1(x, edge_index)
            x = F.relu(x)
            x = self.conv2(x, edge_index)
            return x
    return GCN()

graph_data = build_graph_from_embeddings(features_numpy)
graph_data = graph_data.to(device)
pseudo_labels = pseudo_labels.to(device)

model = create_gcn(all_features.shape[1], 64, 4).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(30):
    model.train()
    optimizer.zero_grad()
    out = model(graph_data.x, graph_data.edge_index)
    loss = F.cross_entropy(out, pseudo_labels)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, GNN Loss: {loss.item():.4f}")

gnn_features = out.detach().cpu().numpy()

# STEP 13: Visualizations
pca = PCA(n_components=2)
pca_result = pca.fit_transform(gnn_features)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=pseudo_labels.cpu(), cmap='viridis')
plt.title('GNN Output Embeddings (PCA)')
plt.xlabel('PCA 1'); plt.ylabel('PCA 2'); plt.show()

tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(features_numpy)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=pseudo_labels.cpu(), cmap='tab10', alpha=0.7)
plt.title('t-SNE of ViT Feature Embeddings')
plt.xlabel('t-SNE 1'); plt.ylabel('t-SNE 2'); plt.grid(True); plt.show()
```

```python
# Install torch_geometric dependencies
!pip install torch_geometric torch_sparse torch_scatter torch_cluster torch_spline_conv -f https://data.pyg.org/whl/torch-2.0.0+cu118.html
import torch
import torch.nn as nn
from torch_geometric.data import Data
from PIL import Image
from torchvision import transforms
from google.colab import files
from transformers import ViTModel, ViTConfig
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Transform setup
vit_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# ViT feature extractor
vit_config = ViTConfig.from_pretrained("google/vit-base-patch16-224-in21k", output_hidden_states=False)
vit_model = ViTModel.from_pretrained("google/vit-base-patch16-224-in21k", config=vit_config).to(device)

# GCN Models
class GCN_Detect(nn.Module):
    def __init__(self, in_features, out_features):
        super(GCN_Detect, self).__init__()
        self.fc1 = nn.Linear(in_features, out_features)
        self.fc2 = nn.Linear(out_features, 2)

    def forward(self, x, edge_index=None):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

class GCN_Classify(nn.Module):
    def __init__(self, in_features, out_features):
        super(GCN_Classify, self).__init__()
        self.fc1 = nn.Linear(in_features, out_features)
        self.fc2 = nn.Linear(out_features, 4)

    def forward(self, x, edge_index=None):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

gcn_detect = GCN_Detect(in_features=768, out_features=256).to(device)
gcn_classify = GCN_Classify(in_features=768, out_features=256).to(device)

# Spread GNN Model
class SpreadGNN(nn.Module):
    def __init__(self, in_channels, hidden_channels):
        super(SpreadGNN, self).__init__()
        self.fc1 = nn.Linear(in_channels, hidden_channels)
        self.fc2 = nn.Linear(hidden_channels, 2)

    def forward(self, x, edge_index=None):
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

# Graph builder
def build_similarity_graph(features, threshold=0.85):
    similarity = cosine_similarity(features)
    edge_index = []

    for i in range(len(features)):
        for j in range(len(features)):
            if i != j and similarity[i, j] > threshold:
                edge_index.append([i, j])

    if not edge_index:
        return torch.empty((2, 0), dtype=torch.long)

    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    return edge_index

# Visualization
def visualize_pollution_spread(embeddings, labels=None):
    pca = PCA(n_components=2)
    reduced = pca.fit_transform(embeddings)

    plt.figure(figsize=(8, 6))
    if labels:
        scatter = plt.scatter(reduced[:, 0], reduced[:, 1], c=labels, cmap="Set1")
        plt.legend(*scatter.legend_elements(), title="Class")
    else:
        plt.scatter(reduced[:, 0], reduced[:, 1], c='blue')

    for i, name in enumerate(image_names):
        plt.annotate(f"{i+1}", (reduced[i, 0], reduced[i, 1]))

    plt.title("Pollution Spread Analysis (PCA)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True)
    plt.show()

# Upload function
def upload_image():
    uploaded = files.upload()
    return next(iter(uploaded))

# Global lists for features
all_features = []
detected_classes = []
image_names = []

# Extract and store features if microplastic detected
def extract_feature_if_microplastic(image_path):
    img = Image.open(image_path).convert("RGB")
    img_tensor = vit_transform(img).unsqueeze(0).to(device)

    # ViT feature extraction
    with torch.no_grad():
        features = vit_model(img_tensor).last_hidden_state[:, 0, :]

    # Detection
    with torch.no_grad():
        detection_output = gcn_detect(features)
        detection_pred = torch.argmax(torch.softmax(detection_output, dim=1), dim=1).item()

    if detection_pred == 0:
        print(f"🟢 No microplastics in {image_path} — Skipping.")
        return

    print(f"✅ Microplastics detected in {image_path} — Feature added.")

    # Classification
    with torch.no_grad():
        class_output = gcn_classify(features)
        class_pred = torch.argmax(torch.softmax(class_output, dim=1), dim=1).item()

    all_features.append(features.squeeze().cpu())
    detected_classes.append(class_pred)
    image_names.append(image_path)

# -------- RUN BELOW TO START UPLOADING IMAGES --------

print("📤 Upload at least 2 images with microplastics for spread analysis:")

for _ in range(2):  # Change this to 3 or 4 to upload more images
    path = upload_image()
    extract_feature_if_microplastic(path)

# -------- PERFORM SPREAD ANALYSIS --------

if len(all_features) >= 2:
    stacked = torch.stack(all_features)
    edge_index = build_similarity_graph(stacked)

    spread_model = SpreadGNN(in_channels=768, hidden_channels=128).to(device)
    spread_model.eval()
    with torch.no_grad():
        embeddings = spread_model(stacked.to(device)).cpu()

    visualize_pollution_spread(embeddings, labels=detected_classes)
else:
    print("❗ Need at least 2 microplastic-positive images for spread analysis.")
```

```python
# STEP 1: Install Required Libraries
!pip install -q transformers torchvision matplotlib torch_geometric

# STEP 2: Imports
import os, glob, zipfile, random
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torch import nn
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset
import torchvision.transforms as transforms
import torchvision
from torchvision import datasets
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from transformers import ViTForImageClassification
import torch.nn.functional as F
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score


# STEP 3: Upload and Extract Dataset
from google.colab import files
uploaded = files.upload()  # Upload train.zip
with zipfile.ZipFile("train.zip", 'r') as zip_ref:
    zip_ref.extractall("train_data")

image_dir = "train_data/train"
print("Path exists:", os.path.exists(image_dir))

# STEP 4: Transforms
transform = transforms.Compose([
    transforms.Resize(32),
    transforms.CenterCrop(32),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# STEP 5: DCGAN Modules (Generator and Discriminator)
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(100, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256), nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128), nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64), nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, input):
        return self.main(input)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.main = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.BatchNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 1, 4, 1, 0), nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)

# Function to Generate and Save Images
def generate_and_save_images(generator, num_images=500, save_dir="synthetic_images"):
    class_dir = os.path.join(save_dir, "class0")
    os.makedirs(class_dir, exist_ok=True)
    generator.eval()
    z = torch.randn(num_images, 100, 1, 1, device=device)
    with torch.no_grad():
        fake_images = generator(z).detach().cpu()
    for i, img_tensor in enumerate(fake_images):
        img = transforms.ToPILImage()(img_tensor)
        img.save(os.path.join(class_dir, f"synthetic_{i}.png"))
    print(f"Saved {num_images} synthetic images to '{class_dir}'")

    # Show 8 samples
    grid = torchvision.utils.make_grid(fake_images[:8], nrow=4, normalize=True)
    plt.figure(figsize=(8, 4))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')
    plt.title("Sample Synthetic Images (DCGAN)")
    plt.show()

# Function to Visualize Progress
def visualize_progress(epoch, generator, fixed_noise, device):
    generator.eval()
    with torch.no_grad():
        fake_images = generator(fixed_noise.to(device)).detach().cpu()
    grid = torchvision.utils.make_grid(fake_images[:8], nrow=4, normalize=True)
    plt.figure(figsize=(8, 4))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')
    plt.title(f"Generated Images at Epoch {epoch}")
    plt.show()

# STEP 6: Prepare Real Dataset for GAN
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
real_dataset = datasets.ImageFolder("train_data", transform=transform)
gan_loader = DataLoader(real_dataset, batch_size=64, shuffle=True)

# STEP 7: Train DCGAN (100 epochs, with progress visualization)
generator = Generator().to(device)
discriminator = Discriminator().to(device)
criterion = nn.BCELoss()
opt_G = torch.optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
opt_D = torch.optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))

epochs = 100
fixed_noise = torch.randn(64, 100, 1, 1, device=device)

for epoch in range(epochs):
    for i, (real_images, _) in enumerate(gan_loader):
        real_images = real_images.to(device)
        b_size = real_images.size(0)
        real_labels = torch.ones(b_size, device=device)
        fake_labels = torch.zeros(b_size, device=device)

        z = torch.randn(b_size, 100, 1, 1, device=device)
        fake_images = generator(z)

        # Train Discriminator
        real_loss = criterion(discriminator(real_images).view(-1), real_labels)
        fake_loss = criterion(discriminator(fake_images.detach()).view(-1), fake_labels)
        d_loss = real_loss + fake_loss
        opt_D.zero_grad(); d_loss.backward(); opt_D.step()

        # Train Generator
        output = discriminator(fake_images).view(-1)
        g_loss = criterion(output, real_labels)
        opt_G.zero_grad(); g_loss.backward(); opt_G.step()

    print(f"Epoch {epoch+1}/{epochs}, D Loss: {d_loss.item():.4f}, G Loss: {g_loss.item():.4f}")

    if epoch % 10 == 0:
        visualize_progress(epoch, generator, fixed_noise, device)

# STEP 8: Generate Synthetic Images
generate_and_save_images(generator, num_images=500)

# STEP 9: Load Real + Synthetic Data
synthetic_dataset = datasets.ImageFolder("synthetic_images", transform=transform)
full_dataset = ConcatDataset([real_dataset, synthetic_dataset])
print("Total images (real + synthetic):", len(full_dataset))

train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_ds, test_ds = random_split(full_dataset, [train_size, test_size])
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False)

# STEP 10: Show Mixed Batch
def show_batch(data_loader):
    images, _ = next(iter(data_loader))
    grid = torchvision.utils.make_grid(images[:8], nrow=4, normalize=True)
    plt.figure(figsize=(10, 5))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')
    plt.show()

show_batch(train_loader)

# STEP 11: ViT Feature Extraction
vit_model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224-in21k')
vit_model.to(device)

vit_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

class ViTDataset(Dataset):
    def __init__(self, dataset):
        self.dataset = dataset
        self.transform = vit_transform

    def __len__(self): return len(self.dataset)

    def __getitem__(self, idx):
        img, _ = self.dataset[idx]
        img = self.transform(transforms.ToPILImage()(img))
        return img

vit_loader = DataLoader(ViTDataset(train_ds), batch_size=32, shuffle=False)

all_features = []
for images in vit_loader:
    images = images.to(device)
    with torch.no_grad():
        outputs = vit_model(images).logits
    all_features.append(outputs.cpu())
all_features = torch.cat(all_features, dim=0)

# STEP 12: KMeans + GCN
features_numpy = all_features.numpy()
kmeans = KMeans(n_clusters=4)
pseudo_labels = kmeans.fit_predict(features_numpy)
print("Silhouette Score:", silhouette_score(features_numpy, pseudo_labels))
pseudo_labels = torch.tensor(pseudo_labels, dtype=torch.long)

def build_graph_from_embeddings(embeddings, k=5):
    nbrs = NearestNeighbors(n_neighbors=k+1, metric='cosine').fit(embeddings)
    _, indices = nbrs.kneighbors(embeddings)
    edge_index = []
    for i in range(len(embeddings)):
        for j in indices[i][1:]:
            edge_index.append([i, j])
    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    return Data(x=torch.tensor(embeddings, dtype=torch.float32), edge_index=edge_index)

def create_gcn(in_channels, hidden_channels, out_channels):
    class GCN(nn.Module):
        def __init__(self):
            super().__init__()
            self.conv1 = GCNConv(in_channels, hidden_channels)
            self.conv2 = GCNConv(hidden_channels, out_channels)

        def forward(self, x, edge_index):
            x = self.conv1(x, edge_index)
            x = F.relu(x)
            x = self.conv2(x, edge_index)
            return x
    return GCN()

graph_data = build_graph_from_embeddings(features_numpy)
graph_data = graph_data.to(device)
pseudo_labels = pseudo_labels.to(device)

model = create_gcn(all_features.shape[1], 64, 4).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(30):
    model.train()
    optimizer.zero_grad()
    out = model(graph_data.x, graph_data.edge_index)
    loss = F.cross_entropy(out, pseudo_labels)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, GNN Loss: {loss.item():.4f}")

gnn_features = out.detach().cpu().numpy()

# STEP 13: Visualizations
pca = PCA(n_components=2)
pca_result = pca.fit_transform(gnn_features)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=pseudo_labels.cpu(), cmap='viridis')
plt.title('GNN Output Embeddings (PCA)')
plt.xlabel('PCA 1'); plt.ylabel('PCA 2'); plt.show()

tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(features_numpy)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=pseudo_labels.cpu(), cmap='tab10', alpha=0.7)
plt.title('t-SNE of ViT Feature Embeddings')
plt.xlabel('t-SNE 1'); plt.ylabel('t-SNE 2'); plt.grid(True); plt.show()
silhouette = silhouette_score(features_numpy, pseudo_labels)
print("Silhouette Score:", silhouette)
_, predicted_labels = torch.max(out, 1)
accuracy = (predicted_labels == pseudo_labels).sum().item() / len(pseudo_labels)
print(f"GCN Accuracy: {accuracy * 100:.2f}%")
from sklearn.metrics import confusion_matrix
import seaborn as sns

confusion = confusion_matrix(pseudo_labels.cpu().numpy(), predicted_labels.cpu().numpy())
plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', xticklabels=range(4), yticklabels=range(4))
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
from sklearn.metrics import classification_report

report = classification_report(pseudo_labels.cpu().numpy(), predicted_labels.cpu().numpy(), target_names=[f'Class {i}' for i in range(4)])
print("Classification Report:")
print(report)
plt.plot(range(epochs), loss_history)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('GCN Training Loss Over Epochs')
plt.show()
```
